```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Airbnb.csv')
```

```python
df.head()
```

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['price']=df['price'].mask(df['price']==0).fillna(df['price'].mean())
```

```python
#Create dummy variables from Room Type since it is a categorical attribute
df=pd.get_dummies(df, columns=['room_type'])
```

```python
from datetime import datetime
date_format = "%Y-%m-%d"
yearEnd = "2019-12-31"
df['Year_End'] = yearEnd 
df['Year_End'] = df['Year_End'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))

df['last_review'] = pd.to_datetime(df['last_review'])

#Calculating the days since last review received for a listing by taking the difference between the date and end of year 2019
df['DaysSinceLastReview'] = df['Year_End'] - df['last_review']
df['DaysSinceLastReview']=df['DaysSinceLastReview'].dt.days
```

```python
len(df.columns)
```

```python
#Most reviews are received within one year, so categorizing the review recency into three categories
Review = pd.cut(df.DaysSinceLastReview, bins=[0,210,365,50000],labels=['InSevenMonths','OneYear','MoreThanOne'])
df.insert(20,'Review Recency', Review)
```

```python
#Check the distribution of the three categories to ensure that the divide is not too skewed
df['Review Recency'].value_counts(normalize=True)
```

```python
#The listings that haven't received a review will have NaN as review recency as a category. 
df=pd.get_dummies(df, columns=['Review Recency'])
```

```python
#The variable "calculated_host_listings_count" is very skewed, so converting it to categorical variable based on volumn using pandas cut
HostList = pd.cut(df.calculated_host_listings_count, bins=[0,1,4,500],labels=['One','OnetoFour','MoreThanfour'])
df.insert(16,'Host Listing', HostList)
```

```python
#Checking the distribution of the three categories
df['Host Listing'].value_counts(normalize=True)
```

```python
df=pd.get_dummies(df, columns=['Host Listing'])
#multicollinearity will be considered in the model
```

```python
#The variable "minimum_nights" is very skewed, so converting it to categorical variable using pandas cut
MiniNight = pd.cut(df.minimum_nights, bins=[0,1,4,1251],labels=['One','TwotoFour','MoreThanfour'])
df.insert(21,'MiniNight', MiniNight)
```

```python
#Checking the distribution of the three categories
df['MiniNight'].value_counts(normalize=True)
```

```python
df=pd.get_dummies(df, columns=['MiniNight'])
#multicollinearity will be considered in the model
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm_model=lm.fit(X_train, y_train)
```

```python
predictions = lm.predict(X_test)
```

```python
predictions
```

```python
plt.scatter(y_test, predictions)
```

```python
#Calculate adj r^2
yhat = lm.predict(X_test)
SS_Residual = sum((y_test-yhat)**2)       
SS_Total = sum((y_test-np.mean(y_test))**2)     
r_squared = 1 - (float(SS_Residual))/SS_Total
adjusted_r_squared = 1 - (1-r_squared)*(len(y)-1)/(len(y)-X.shape[1]-1)

adjusted_r_squared
```

```python
lm.score(X_test, y_test)
```

```python
from sklearn.cluster import KMeans
from collections import Counter
```

```python
Coordinate=df.loc[:,['latitude','longitude']].values
```

```python
#Initial k mean model
id_n=1
kmeans = KMeans(n_clusters=id_n, random_state=0).fit(Coordinate)
```

```python
from scipy.spatial.distance import cdist 
```

```python
#Choose the number of cluster using the elbow method
distortions = [] 
inertias = [] 
mapping1 = {} 
mapping2 = {} 
K = range(1,15) 
  
for id_n in K: 
    #Building and fitting the model 
    kmeans = KMeans(n_clusters=id_n).fit(Coordinate) 
    kmeans.fit(Coordinate)     
      
    distortions.append(sum(np.min(cdist(Coordinate, kmeans.cluster_centers_, 
                      'euclidean'),axis=1)) / Coordinate.shape[0]) 
    inertias.append(kmeans.inertia_) 
  
    mapping1[id_n] = sum(np.min(cdist(Coordinate, kmeans.cluster_centers_, 
                 'euclidean'),axis=1)) / Coordinate.shape[0] 
    mapping2[id_n] = kmeans.inertia_ 
```

```python
plt.plot(K, distortions, 'bx-') 
plt.xlabel('Values of K') 
plt.ylabel('Distortion') 
plt.title('The Elbow Method using Distortion') 
plt.show() 
#The graph shows that the distortion is continuing to decrease
#for the sake of model performance, the model is choosing k=8
```
